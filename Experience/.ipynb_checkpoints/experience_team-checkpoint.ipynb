```python
import pandas as pd
from datetime import timedelta
from datetime import datetime

# Declare variable here
startTime = datetime(2023,2,12,14,32,6)
gen_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
correct = "correct"
question_key = "問題Question"
question = "question"
```

```python
#E001 前九廣鐵路鐘樓
cp01 = pd.read_csv('./exp_record/01.csv')
cp01.rename(columns={question_key:question},inplace=True)
cp01['CP01']=0
cp01.loc[cp01.question == 'correct', 'CP01'] = 50
cp01.loc[cp01.question != 'correct', 'CP01'] = 0
cp01.drop(labels=["ObjectID","CreationDate","GlobalID","Editor","Creator", "x","y", question, "EditDate"], axis=1, inplace=True)
cp01 = cp01.drop_duplicates(subset=['Team Name (Hide)'])

#E002 戲曲中心
cp02 = pd.read_csv('./exp_record/02.csv')
cp02.rename(columns={question_key:question},inplace=True)
cp02['CP02']=0
cp02.loc[cp02.question == correct, 'CP02'] = 50
cp02.loc[cp02.question != correct, 'CP02'] = 0
cp02.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp02 = cp02.drop_duplicates(subset=['Team Name (Hide)'])

#E003 香港女童軍新總部大樓地盤
cp03 = pd.read_csv('./exp_record/03.csv')
cp03.rename(columns={question_key:question},inplace=True)
cp03['CP03']=0
cp03.loc[cp03.question == correct, 'CP03'] = 50
cp03.loc[cp03.question != correct, 'CP03'] = 0
cp03.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp03 = cp03.drop_duplicates(subset=['Team Name (Hide)'])

#E004 星光大道
cp04 = pd.read_csv('./exp_record/04.csv')
cp04.rename(columns={question_key:question},inplace=True)
cp04['CP04']=0
cp04.loc[cp04.question == correct, 'CP04'] = 50
cp04.loc[cp04.question != correct, 'CP04'] = 0
cp04.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp04 = cp04.drop_duplicates(subset=['Team Name (Hide)'])

#E005 聖安德烈堂
cp05 = pd.read_csv('./exp_record/05.csv')
cp05.rename(columns={question_key:question},inplace=True)
cp05['CP05']=0
cp05.loc[cp05.question == correct, 'CP05'] = 10
cp05.loc[cp05.question != correct, 'CP05'] = 0
cp05.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp05 = cp05.drop_duplicates(subset=['Team Name (Hide)'])

#E006 文物探知館
cp06 = pd.read_csv('./exp_record/06.csv')
cp06.rename(columns={question_key:question},inplace=True)
cp06['CP06']=0
cp06.loc[cp06.question == correct, 'CP06'] = 10
cp06.loc[cp06.question != correct, 'CP06'] = 0
cp06.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp06 = cp06.drop_duplicates(subset=['Team Name (Hide)'])

#E007 K11 Art Mall 
cp07 = pd.read_csv('./exp_record/07.csv')
cp07.rename(columns={question_key:question},inplace=True)
cp07['CP07']=0
cp07.loc[cp07.question == correct, 'cp07(Score)'] = 10
cp07.loc[cp07.question != correct, 'cp07(Score)'] = 0
cp07.drop(["ObjectID", "CreationDate","GlobalID","EditDate","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp07 = cp07.drop_duplicates(subset=['Team Name (Hide)'])

#E008 九龍清真寺
cp08 = pd.read_csv('./exp_record/08.csv')
cp08.rename(columns={question_key:question},inplace=True)
cp08['CP08']=0
cp08.loc[cp08.question == correct, 'CP08'] = 10
cp08.loc[cp08.question != correct, 'CP08'] = 0
cp08.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp08 = cp08.drop_duplicates(subset=['Team Name (Hide)'])

#E009 九龍佑寧堂
cp09 = pd.read_csv('./exp_record/09.csv')
cp09.rename(columns={question_key:question},inplace=True)
cp09['CP09']=0
cp09.loc[cp09.question == correct, 'CP09'] = 10
cp09.loc[cp09.question != correct, 'CP09'] = 0
cp09.drop(labels=["ObjectID","CreationDate","EditDate","GlobalID","Creator","Editor", "x","y", question], axis=1, inplace=True)
cp09 = cp09.drop_duplicates(subset=['Team Name (Hide)'])

#E010 香港女童軍總會
cp10 = pd.read_csv('./exp_record/10.csv')
cp10.rename(columns={question_key:question},inplace=True)
cp10.rename(columns={"EditDate":"FinishTime"},inplace=True)
cp10.drop(labels=["ObjectID","CreationDate","GlobalID","Editor", "Creator", "x","y",question], axis=1, inplace=True)
cp10 = cp10.drop_duplicates(subset=['Team Name (Hide)'])
```

```python
df_list = [cp01,cp02,cp03,cp04,cp05,cp06,cp07,cp08,cp09,cp10]
master = df_list[0]
for df in df_list[1:]:
    master = pd.merge(master, df, on='Team Name (Hide)', how="outer")
    
master = master.drop_duplicates(subset=['Team Name (Hide)'])

master['Total Time']=''
master['IsComplete']=False
master['Total Score']=0

pd.options.display.float_format = '{:,.0f}'.format

master.rename(columns={"Team Name (Hide)":"Team Name"},inplace=True)
master.loc[master.FinishTime.notnull(), "IsComplete"]=True
###Sumation for length of passing
master.loc[master.IsComplete == True, 'Total Score']= master[["CP01","CP02","CP03","CP04","CP05","CP06","CP07","CP08","CP09"]].sum(1)

master.FinishTime = pd.to_datetime(master.FinishTime)
master['FinishTime'] = master['FinishTime']+ timedelta(hours=8)
master.loc[master.IsComplete == True,'Total Time']=master["FinishTime"]-startTime
master['Total Time'] = master['Total Time'].astype(str).str.replace('0 days ', '')
master['Seconds'] = ''
master['Seconds'] = master['Total Time'] / pd.Timedelta(seconds=1)
master.loc[master.Seconds>7200, "IsComplete"]=False
master.loc[(master['CP02'] + master['CP04'] == 0), "IsComplete"]=False
master.loc[pd.isnull(master['CP02']) & pd.isnull(master['CP04']), "IsComplete"]=False
master.sort_values(by=["IsComplete",'Total Score', 'Total Time', 'Team Name'], ascending=[False, False, True, True], inplace=True)
master.drop(labels=['Seconds'], axis=1, inplace=True)

master = master.reset_index(drop=True)
master.index += 1
master.columns.name = 'Position'
master['Position'] = master.index
master = master.fillna('abs')
order = ["Position","Team Name","CP01","CP02","CP03","CP04","CP05","CP06","CP07", "CP08","CP09","FinishTime","IsComplete","Total Time","Total Score"]
master= master[order]

public_list = master.copy()
public_list = public_list[public_list['Team Name'].astype(str).str.startswith('EXP')]
public_list.drop(labels=['FinishTime', 'CP01','CP02','CP03','CP04','CP05','CP06','CP07','CP08','CP09'], axis=1, inplace=True)
order_n = ["Position","Team Name","Total Time","Total Score","IsComplete"]
public_list= public_list[order_n]

master.sort_values(by=['Team Name'], ascending=[True], inplace=True)
master.to_csv('master_experience.csv',sep=',', header=True, index=False)
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("experience_template.html", column_names=public_list.columns.values, row_data=list(public_list.values.tolist()), zip=zip, date=gen_time)

app.run()
```
