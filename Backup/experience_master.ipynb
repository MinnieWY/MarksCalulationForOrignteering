```python
df_list = [cp01,cp02,cp03,cp04,cp05,cp06,cp07,cp08,cp09,cp10,cp11]
master = df_list[0]
for df in df_list[1:]:
    master = pd.merge(master, df, on='Creator', how="outer")
    
master = master.drop_duplicates(subset=['Creator'])

master['Total Time']=''
master['IsComplete']=False
master['IsBooth']= False
master['IsCheckPoint']= False
master['Total Score']=0
master['Achievement (Y/N)']=False

pd.options.display.float_format = '{:,.0f}'.format

master.rename(columns={"Creator":"Team Name"},inplace=True)
master.loc[master.FinishTime.notnull(), "IsComplete"]=True
master.loc[(master['CP02(Visit)']==True) | (master['CP04(Visit)']==True), 'IsBooth'] = True
master.loc[(master['CP02(Visit)']==False) | (master['CP04(Visit)']==False), 'CP02'] = abs
###Sumation for length of passing
master.loc[master.IsComplete == True, 'Total Score']= master[["CP01","CP02","CP03","CP04","CP05","CP06","CP07","CP08",'CP09','CP10',]].sum(1)
master.loc[master['Total Score'] >= 20, "IsCheckPoint"]=True
master.loc[(master['IsBooth'] == True) & (master['IsCheckPoint'] == True), "Achievement (Y/N)"]=True

master.FinishTime = pd.to_datetime(master.FinishTime)
master['FinishTime'] = master['FinishTime']+ timedelta(hours=8)
master.loc[master.IsComplete == True,'Total Time']=master["FinishTime"]-startTime
master['Total Time'] = master['Total Time'].astype(str).str.replace('0 days ', '')
master['Seconds'] = ''
master['Seconds'] = master['Total Time'] / pd.Timedelta(seconds=1)
master.loc[master.Seconds>7200, "IsComplete"]=False
master.sort_values(by=["IsComplete",'Achievement (Y/N)','Total Score', 'Total Time', 'Team Name'], ascending=[False, False, False, True, True], inplace=True)
master.drop(labels=['Seconds'], axis=1, inplace=True)

master = master.reset_index(drop=True)
master.index += 1
master.columns.name = 'Position'
master = master.fillna('abs')
master['Achievement (Y/N)'] = master['Achievement (Y/N)'].map({True: 'Y', False: 'N'})
master['CP02'] = master['CP02'].map({True: 'Y', False: 'N'})
master['CP04'] = master['CP04'].map({True: 'Y', False: 'N'})
master['IsBooth'] = master['IsBooth'].map({True: 'Y', False: 'N'})
master['IsCheckPoint'] = master['IsCheckPoint'].map({True: 'Y', False: 'N'})
order = ["Team Name","CP01","CP02(Visit)","CP02","CP03","CP04(Visit)","CP04","CP05","CP06","CP07", "CP08",'CP09','CP10',"FinishTime","IsComplete","Total Time","IsBooth","IsCheckPoint","Total Score", "Achievement (Y/N)"]
master= master[order]
public_list = master.copy()
public_list.drop(labels=['FinishTime', 'CP01','CP02(Visit)','CP02','CP03','CP04(Visit)','CP04','CP05','CP06','CP07','CP08','CP09','CP10', 'IsBooth', 'IsCheckPoint', 'IsComplete'], axis=1, inplace=True)

master.sort_values(by=['Team Name'], ascending=[True], inplace=True)
master.to_csv('master_experience.csv',sep=',', header=True, index=False)
```
