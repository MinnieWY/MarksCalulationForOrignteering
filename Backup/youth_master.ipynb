```python
def _draw_as_table(df, pagesize):
    fig, ax = plt.subplots(constrained_layout=True)
    fig.patch.set_visible(False)
    ax.axis('tight')
    ax.axis('off')
    ax.set_title("Youth Team Result", fontsize=12)
    the_table = ax.table(cellText=df.values,
                        rowLabels=df.index,
                        colLabels=df.columns,
                        colColours=['lightblue']*len(df.columns),
                        loc='center')
    the_table.auto_set_column_width(col=list(range(len(df.columns))))
    return fig
```

```python
df_list = [cp01,cp02,cp03,cp04,cp05,cp06,cp07,cp08,cp09,cp10,cp11,cp12,cp13,cp14,cp15,cp16]
master = df_list[0]
for df in df_list[1:]:
    master = pd.merge(master, df, on='Creator', how="outer")
master['Total Time']=''
master['IsComplete']=False
master['Total Score']=0

pd.options.display.float_format = '{:,.0f}'.format

master.rename(columns={"Creator":"Team Name"},inplace=True)
master.loc[master.FinishTime.notnull(), "IsComplete"]=True
master.loc[master.IsComplete == True, 'Total Score']= master[["CP01","CP02","CP03","CP04","CP05","CP06","CP07", "CP08","CP09","CP10","CP11","CP12","CP13","CP14","CP15"]].sum(1)

master.FinishTime = pd.to_datetime(master.FinishTime)
master['FinishTime'] = master['FinishTime']+ timedelta(hours=8)
master.loc[master.IsComplete == True,'Total Time']=master["FinishTime"]-startTime
master['Total Time'] = master['Total Time'].astype(str).str.replace('0 days ', '')
master['Seconds'] = ''
master['Seconds'] = master['Total Time'] / pd.Timedelta(seconds=1)
master.sort_values(by=['IsComplete' ,'Total Score', 'Total Time', 'Team Name'], ascending=[False, False, True, True], inplace=True)

master = master.reset_index(drop=True)
master.index += 1
master['Position'] = master.index
master.drop(labels=['Seconds'], axis=1, inplace=True)
master = master.fillna('abs')
order = ["Position","Team Name","CP01","CP02","CP03","CP04","CP05","CP06","CP07", "CP08","CP09","CP10","CP11","CP12","CP13","CP14","CP15","FinishTime","IsComplete","Total Time","Total Score"]
master= master[order]

public_list = master.copy()
public_list.drop(labels=["Position",'FinishTime',"CP01","CP02","CP03","CP04","CP05","CP06","CP07", "CP08","CP09","CP10","CP11","CP12","CP13","CP14","CP15", "IsComplete"], axis=1, inplace=True)

master.sort_values(by=['Team Name'], ascending=[True], inplace=True)
master.to_csv('master_youth.csv',sep=',', header=True, index=False)

 
with PdfPages("public_youth.pdf") as pdf:
  fig = _draw_as_table(public_list, (10.5,8))
  pdf.savefig(fig, bbox_inches='tight')
  plt.close()
```
